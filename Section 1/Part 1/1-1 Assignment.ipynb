```python
import numpy as np

with open('test_data.csv') as f:
    data = np.loadtxt(f, delimiter=',')

input_data = data[:,0]  # first column
output_data = data[:,2]  # third column
```

```python
%matplotlib inline
from matplotlib import pyplot as plt

plt.scatter(input_data, output_data)
plt.show()
```

```python
from sklearn import linear_model

# Training inputs and outputs have to be reshaped to 2D matrices.
train_input = np.array(input_data).reshape(-1, 1)
train_output = np.array(output_data).reshape(-1, 1)

regr = linear_model.LinearRegression()
regr.fit(train_input, train_output)
predictions = regr.predict(train_input)
```

```python
plt.xlabel('x')
plt.ylabel('y')
plt.title('Assignment 1')

plt.scatter(input_data, output_data)
plt.plot(train_input, predictions)
plt.show()
```
